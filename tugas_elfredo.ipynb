```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv(r'company_sales_data.csv')
data.head(12)
```

```python
data.columns
```

```python
plt.plot('month_number', 'bathingsoap', data=data, label='Bathingsoap Sales Data', marker='o', linewidth=3)
```

```python
monthList  = data ['month_number'].tolist()
bathingsoapSalesData   = data ['bathingsoap'].tolist()
plt.bar(monthList, bathingsoapSalesData, width= 0.35, label = 'Face Cream sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Penjualan Toko EL')
```

```python
plt.figure(figsize=(7,5))
plt.plot('month_number', 'total_profit', data=data)
plt.xticks(data['month_number'].tolist())
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot('month_number', 'total_profit', data=data, ls='--', c='green', marker='o', markerfacecolor='blue',
         label='Profit data of last year', lw=3)
plt.xticks(data['month_number'].tolist())
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc = 'lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data untuk setiap diagram batang
plt.figure(figsize=(11,5))
monthList = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']
faceCreamSalesData = [2500,2630,2140,3400,3600,2760,2980]
faceWashSalesData = [1500,1200,1340,1130,1740,1555,1120]
toothPasteSalesData = [5200,5100,4550,5870,4560,4890,4780]
bathingSoapSalesData = [9200,6100,9550,8870,7760,7490,8980]
shampooSalesData = [1200,2100,3550,1870,1560,1890,1780]
moisturizerSalesData = [1500,1200,1340,1130,1740,1555,1120]

# Lebar rata-rata untuk setiap diagram batang
width = 0.15

# Membuat posisi sumbu x
x = np.arange(len(monthList))

# Membuat diagram batang
plt.bar(x - 2*width, faceCreamSalesData, width=width, label='Face Cream')
plt.bar(x - width, faceWashSalesData, width=width, label='Face Wash')
plt.bar(x, toothPasteSalesData, width=width, label='Tooth Paste')
plt.bar(x + width, bathingSoapSalesData, width=width, label='Bathing Soap')
plt.bar(x + 2*width, shampooSalesData, width=width, label='Shampoo')
plt.bar(x + 3*width, moisturizerSalesData, width=width, label='moisturizer')


# Menambahkan label sumbu x dan sumbu y serta judul
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.title('Sales Data')
plt.xticks(x, monthList)  # Mengatur label sumbu x sesuai dengan bulan
plt.legend()

# Menampilkan diagram
plt.show()
```
